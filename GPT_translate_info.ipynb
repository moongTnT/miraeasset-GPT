```python
import os, json

with open('conf.json', 'r') as f:
    json_data = json.load(f)
    
os.environ['OPENAI_API_KEY'] = json_data['API_KEY']
```

```python
from langchain.prompts import ChatPromptTemplate
from langchain.prompts.chat import SystemMessage, HumanMessagePromptTemplate

template = ChatPromptTemplate.from_messages(
    [
        SystemMessage(
            content=(
                "You are a helpful translate assistant that re-writes the user's English text to Korean."
            )
        ),
        HumanMessagePromptTemplate.from_template("{text}"),
    ]
)

from langchain.chat_models import ChatOpenAI

llm = ChatOpenAI(model='gpt-3.5-turbo')
llm(template.format_messages(text='i dont like eating tasty things.'))
```

```python
import os, json

json_list = os.listdir('etf_subcategory_infos')

for json_fname in json_list:
    with open(f'etf_subcategory_infos/{json_fname}', 'r') as f:
        json_data = json.load(f)
        
    print()
    print(json_fname)
    print()
        
    for i, subcategory in enumerate(json_data['subcategory_lists']):
        
        title = subcategory['title']
        
        description = subcategory['description']
        
        title_kr = llm(template.format_messages(text=title)).content
        
        subcategory['title_kr'] = title_kr
        
        if description != "":
            
            description_kr = llm(template.format_messages(text=description)).content
            
            subcategory['description_kr'] = description_kr
            
        json_data['subcategory_lists'][i] = subcategory
        
    print(json_data)
        
    with open(f'./etf_subcategory_infos/{json_fname}', 'w') as outfile:
        json.dump(json_data, outfile, indent='\t', ensure_ascii=False)
```
