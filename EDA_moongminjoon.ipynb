# EDA

## 해외 ETF 사용자의 연령대 분포 비교

```python
df['ETF_TRADING_YN_NEW'] = df[['DMETF_FST_BUY_YM', 'OVETF_FST_BUY_YM']].apply(lambda row: int(row.notna().any()), axis=1)

age_etf_counts = df.groupby('AGE_TCD')['ETF_TRADING_YN_NEW'].sum()


age_total_counts = df['AGE_TCD'].value_counts().sort_index()


age_all_etf_percentage = (age_etf_counts / age_total_counts * 100).fillna(0)

age_all_etf_percentage
```

```python
# Import the necessary library
import matplotlib.pyplot as plt

# Color palette
color_palette = ['#F0B26B', '#CB6015', '#AD624E', '#84888B', '#8DC8E8', '#0086B8', '#00A9CE', '#7E9FC3', '#7E9FC3']

# Create the bar graph
fig, ax = plt.subplots()
bars = plt.bar(age_all_etf_percentage.index, age_all_etf_percentage, color=color_palette[:len(age_all_etf_percentage)], width=5)

# Add labels and title
plt.xlabel('연령대')
plt.ylabel('ETF 거래 비율 (%)')
plt.title('연령대별 국내+해외 ETF 거래 비율')

# Annotate with the actual values
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, f"{yval:.1f}%", va='bottom', ha='center')  # va: vertical alignment, ha: horizontal alignment

# Show the plot
plt.show()
```

20, 30대가 해외 ETF를 많이 사용한다는 것을 알 수 있다. 

```python
# Domestic, Overseas, and both ETF trading
df['DMETF_TRADING_YN'] = df['DMETF_FST_BUY_YM'].notna().astype(int)
df['OVETF_TRADING_YN'] = df['OVETF_FST_BUY_YM'].notna().astype(int)
df['BOTH_ETF_TRADING_YN'] = df[['DMETF_FST_BUY_YM', 'OVETF_FST_BUY_YM']].apply(lambda row: int(row.notna().all()), axis=1)

# 연령대별 그룹짓기
age_dm_counts = df.groupby('AGE_TCD')['DMETF_TRADING_YN'].sum()
age_ov_counts = df.groupby('AGE_TCD')['OVETF_TRADING_YN'].sum()
age_both_counts = df.groupby('AGE_TCD')['BOTH_ETF_TRADING_YN'].sum()

# 퍼센트 게산
age_dm_percentage = (age_dm_counts / age_total_counts * 100).fillna(0)
age_ov_percentage = (age_ov_counts / age_total_counts * 100).fillna(0)
age_both_percentage = (age_both_counts / age_total_counts * 100).fillna(0)
```

```python

# Plotting
fig, ax = plt.subplots(figsize=(15, 6))  

x = np.arange(len(age_total_counts.index))

width = 0.2

bars1 = plt.bar(x - width, age_dm_percentage, width, label='국내 ETF', color=color_palette[0])
bars2 = plt.bar(x, age_ov_percentage, width, label='해외 ETF', color=color_palette[1])
bars3 = plt.bar(x + width, age_both_percentage, width, label='모두', color=color_palette[2])

plt.xlabel('연령대')
plt.ylabel('ETF 거래 비율 (%)')
plt.title('연령대별 국내, 해외, 국내+해외 ETF 거래 비율')

plt.xticks(x, age_total_counts.index) 
plt.legend()

for bars in [bars1, bars2, bars3]:
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2.0, yval, f"{yval:.1f}%", va='bottom', ha='center')

plt.show()





```

국내 ETF만 하는 사람들이 많다

## 연령대별 주식 매수, 매도 비율


```python
summed_df = pd.DataFrame()

prefix_list = ['BUY_CNT', 'SEL_CNT', 'BUY_ITM_CNT', 'SEL_ITM_CNT', 'BUY_AMT', 'SEL_AMT', 'STK_IN', 'STK_OUT', 'MNY_IN', 'MNY_OUT']

for prefix in prefix_list:
    filtered_columns = [col for col in df.columns if prefix in col]
    filtered_df = df[filtered_columns]
    
    df[f"{prefix}_SUM"] = filtered_df.sum(axis=1)

summed_columns = [f"{prefix}_SUM" for prefix in prefix_list]
summed_df = df[['AGE_TCD'] + summed_columns]

grouped_df = summed_df.groupby('AGE_TCD').sum()

print(grouped_df)
```

```python
grouped_df['BUY_RATIO'] = grouped_df['BUY_CNT_SUM'] / (grouped_df['BUY_CNT_SUM'] + grouped_df['SEL_CNT_SUM'])
grouped_df['SELL_RATIO'] = grouped_df['SEL_CNT_SUM'] / (grouped_df['BUY_CNT_SUM'] + grouped_df['SEL_CNT_SUM'])

print(grouped_df[['BUY_RATIO', 'SELL_RATIO']])
```

```python

import pandas as pd
import matplotlib.pyplot as plt


grouped_df = df_no_null.groupby(['AGE_TCD', 'MT_EP_EXIST_YN']).size().reset_index(name='Counts')

total_counts = grouped_df.groupby('AGE_TCD')['Counts'].transform('sum')
grouped_df['Percentage'] = (grouped_df['Counts'] / total_counts) * 100

data_N = grouped_df[grouped_df['MT_EP_EXIST_YN'] == 'N']
data_Y = grouped_df[grouped_df['MT_EP_EXIST_YN'] == 'Y']

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.35
index = range(len(data_N['AGE_TCD']))

bar1 = plt.bar(index, data_N['Counts'], bar_width, label='No Management Employee (N)', color='b')
bar2 = plt.bar([i + bar_width for i in index], data_Y['Counts'], bar_width, label='With Management Employee (Y)', color='r')

plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Counts', fontsize=12)
plt.title('Existence of Management Employee by Age Group', fontsize=14)
plt.xticks([i + bar_width / 2 for i in index], data_N['AGE_TCD'])

plt.legend()

plt.tight_layout()
plt.show()
```
